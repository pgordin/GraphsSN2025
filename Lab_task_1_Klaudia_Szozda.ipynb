Added a function print_graph from labs in order to print a graph in a nice form

```python
def print_graph(graph):
  """
  Prints a graph given by adjacency list
  """
  for v in graph:
    print(v, ":", end=' ', sep = "")
    for u in graph[v]:
      print(u, end=' ')
    print()
```

## Task 1

Write a function random_graph(n, p), which returns a random graph in G(n, p) model — n vertices, an existence of each edge is independent and holds with probability p.

```python
import random
def random_graph(n, p):
  # Initialize empty graph
  graph = {i: [] for i in range(n)}

  # Iterate over all pairs of distinct vertices
  for i in range(n):
      for j in range(i + 1, n):
          if random.random() < p:
              # Add undirected edge (i, j)
              graph[i].append(j)
              graph[j].append(i)

  return graph
```

```python
# use of code
n = 6
p = 0.35
ex_graph = random_graph(n, p)
print_graph(ex_graph)
```

## Task 2

Write a function graph_to_matrix(graph), which converts a graph given by neighbour list into matrix representation and returns a matrix and a sequence of vertices.


```python
def graph_to_matrix(graph):
  # Get list of vertices
  vertices = list(graph.keys())

  # Initialize empty matrix
  matrix = [[0 for _ in vertices] for _ in vertices]

  # Fill matrix with edge values
  for i in range(len(vertices)):
      for j in graph[vertices[i]]:
          matrix[i][j] = 1

  return matrix, vertices
```

# Task 3

Write a function matrix_to_graph(vertices, matrix) which converts in the opposite direction — convertiing a matrix (and a sequence of vertex labels) into a form of neighbour list (python dictionary).


```python
def matrix_to_graph(vertices, matrix):
  # Initialize empty graph
  graph = {i: [] for i in vertices}

  # Fill graph with edges
  for i in range(len(vertices)):
      for j in range(len(vertices)):
          if matrix[i][j] == 1:
              graph[vertices[i]].append(vertices[j])

  return graph
```

## Task 4

Write a function cycle(n), which return a cycle on n vertices.

```python
def cycle(n):
  "returns a cycle on n vertices"
  # we have to assume that n>0
  if n == 0:
    raise ValueError("n must be greater than 0")

  graph = {i: [] for i in range(n)}
  for i in range(n):
    v = i
    u = (i+1)%n
    if v not in graph[u]:
        graph[u].append(v)
    if u not in graph[v]:
        graph[v].append(u)
  return graph
```
